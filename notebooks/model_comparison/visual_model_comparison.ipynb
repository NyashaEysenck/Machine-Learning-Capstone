```python
import matplotlib.pyplot as plt
import numpy as np

# Given data
models_rmse = ['KNN', 'NMF', 'Base Linear Regression', 'RandomForestRegressor', 'RecommenderNet']
rmse_values = [0.8827, 1.2199, 0.6868, 0.6623, 0.1196]

models_accuracy = ['Logistic Regression', 'Decision Tree', 'Bagging', 'Boosting']
accuracy_values = [0.33446916120183445, 0.3359478805023359, 0.33704084694183706, 0.33584072692983585]

# Sorting models based on RMSE and accuracy values
sorted_models_rmse = [x for _, x in sorted(zip(rmse_values, models_rmse))]
sorted_rmse_values = sorted(rmse_values)

sorted_models_accuracy = [x for _, x in sorted(zip(accuracy_values, models_accuracy))]
sorted_accuracy_values = sorted(accuracy_values)

# Plotting the bar chart for RMSE
plt.figure(figsize=(10, 6))
plt.barh(np.arange(len(sorted_models_rmse)), sorted_rmse_values, color='salmon', edgecolor='black')
plt.yticks(np.arange(len(sorted_models_rmse)), sorted_models_rmse, fontsize=12)
plt.xlabel('RMSE', fontsize=14)
plt.title('Performance of Models (Ordered by RMSE)', fontsize=16)
plt.gca().invert_yaxis()  # Invert y-axis to display the highest RMSE at the top
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Plotting the bar chart for accuracy with adjusted scale
plt.figure(figsize=(10, 6))
plt.barh(np.arange(len(sorted_models_accuracy)), sorted_accuracy_values, color='lightblue', edgecolor='black')
plt.yticks(np.arange(len(sorted_models_accuracy)), sorted_models_accuracy, fontsize=12)
plt.xlabel('Accuracy', fontsize=14)
plt.title('Performance of Models (Ordered by Accuracy)', fontsize=16)
plt.xlim(0.334, 0.338)  # Adjusting scale to highlight small differences
plt.gca().invert_yaxis()  # Invert y-axis to display the highest accuracy at the top
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
